```python
import pandas as pd
import numpy as np
```

```python
#Danish data
DK1 = pd.read_csv('data/Denmark/Raw/P05_07_DK.csv')
DK2 = pd.read_csv('data/Denmark/Raw/P07_11_DK.csv')
DK3 = pd.read_csv('data/Denmark/Raw/P11_15_DK.csv')
DK4 = pd.read_csv('data/Denmark/Raw/P15_19_DK.csv')
DK5 = pd.read_csv('data/Denmark/Raw/P19_22_DK.csv')
DK6 = pd.read_csv('data/Denmark/Raw/P22_present_DK.csv')
df_DK = [DK1,DK2,DK3,DK4,DK5,DK6]

congresses = ['095', '096', '097','098', '099', '100', '101', '102', '103','104',
              '105', '106', '107','108', '109', '110', '111', '112','113', '114',
              '115', '116','117','118']

dfs = [pd.read_csv(f"Data/USA/Filtered/H{congress}_filtered_USA_votes.csv") for congress in congresses]

ak = pd.read_csv('data/Denmark/Danish_actor.csv')
```

```python
#average amount of votes

voteid_unique_DK = set()
personid_unique_DK = set()
per = 0
for i, df in enumerate(df_DK):
    
    afstemning_col = None
    for col in df.columns:
        if col.lower() == 'afstemningid':
            afstemning_col = col
            break
    
    personid_col = None
    for col in df.columns:
        if col.lower() == 'aktørid':
            personid_col = col
            break
    
    

    unique_ids1 = set(df[afstemning_col].unique())
    unique_ids2 = set(df[personid_col].unique())
    
    per += len(unique_ids2)
    print(f"Dataframe {i+1}: {len(unique_ids1)} unique afstemningid values")
    

    voteid_unique_DK.update(unique_ids1)
    personid_unique_DK.update(unique_ids2)

# Print the total count of unique IDs across all dataframes
print(f"\nTotal unique afstemningid values across all dataframes: {len(voteid_unique_DK)/6}")
print(f"\nTotal unique afstemningid values across all dataframes: {per/6}")
```

```python

voteid_unique_US = set()
personid_unique_US = set()

votes = 0
uniq = 0

for i, df in enumerate(dfs):
    
    afstemning_col = None
    for col in df.columns:
        if col.lower() == 'rollnumber':
            afstemning_col = col
            break
    
    personid_col = None
    for col in df.columns:
        if col.lower() == 'icpsr':
            personid_col = col
            break
    

    unique_ids1 = set(df[afstemning_col].unique())
    unique_ids2 = set(df[personid_col].unique())
    
    votes += len(unique_ids1)
    uniq += len(unique_ids2)


    voteid_unique_US.update(unique_ids1)
    personid_unique_US.update(unique_ids2)

# Print the total count of unique IDs across all dataframes
print(f"\ average votes across all periods: {votes/24}")
print(f"\ average representatives across all periods: {uniq/24}")
```

```python
all_unique_ids_list = list(unique_ids2)
        
# Create a dataframe from the list of unique IDs
unique_ids_df = pd.DataFrame({'id': all_unique_ids_list})

# Merge with the ak dataframe to get the corresponding 'navn' and 'typeid'
result = pd.merge(unique_ids_df, ak[['id', 'navn', 'typeid']], on='id', how='left')

# Display the results
print("\nSample of unique IDs with their corresponding 'navn' and 'typeid':")
print(result.head(10))


# Count nulls (IDs that weren't found in the ak dataframe)
null_count = result['navn'].isna().sum()
if null_count > 0:
    print(f"\nWarning: {null_count} IDs from voting data not found in the ak dataframe")

```
