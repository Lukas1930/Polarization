```python
import requests
import pandas as pd

# Base API URL
base_url = "https://oda.ft.dk/api/Afstemning?$filter=year(opdateringsdato) ge 2019 and year(opdateringsdato) le 2025"

# Pagination settings
batch_size = 100  # API limit
skip = 0  # Start at first record
all_data = []  # Store all results

while True:
    # Construct paginated URL
    url = f"{base_url}&$top={batch_size}&$skip={skip}"
    # Fetch data
    response = requests.get(url, headers={"Accept": "application/json"})
    
    # Check for valid response
    if response.status_code != 200:
        print(f"Error: {response.status_code}")
        break  # Stop if there's an issue
    
    # Parse JSON response
    data = response.json()
    print(data)
    # If no more data, break loop
    if len(data['value'])==0:
        break
    
    # Append data to list
    all_data.extend(data['value'])
    print(skip)
    # Move to next batch
    skip += batch_size

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Filter for years 2020-2024
df.to_csv("dansk_data.csv")
```
