```python
import requests
import pandas as pd

# Base API URL
base_url = "https://oda.ft.dk/api/Aktør?"

# Pagination settings
batch_size = 100  # API limit
skip = 0  # Start at first record
all_data = []  # Store all results

while True:
    # Construct paginated URL
    url = f"{base_url}&$top={batch_size}&$skip={skip}"
    # Fetch data
    response = requests.get(url, headers={"Accept": "application/json"})
    
    # Check for valid response
    if response.status_code != 200:
        print(f"Error: {response.status_code}")
        break  # Stop if there's an issue
    
    # Parse JSON response
    data = response.json()
    print(data)
    # If no more data, break loop
    if len(data['value'])==0:
        break
    
    # Append data to list
    all_data.extend(data['value'])
    print(skip)
    # Move to next batch
    skip += batch_size

# Convert to DataFrame
df = pd.DataFrame(all_data)

```

```python
import requests
import pandas as pd

# Base API URL for afstemning endpoint
base_url = "https://oda.ft.dk/api/Afstemning?"

# Pagination settings
batch_size = 100  # API limit
skip = 0  # Start at first record
all_data = []  # Store all results

while True:
    # Construct paginated URL
    url = f"{base_url}&$top={batch_size}&$skip={skip}&$select=id,mødeid"
    
    # Fetch data
    response = requests.get(url, headers={"Accept": "application/json"})
    
    # Check for valid response
    if response.status_code != 200:
        print(f"Error: {response.status_code}")
        break  # Stop if there's an issue
    
    # Parse JSON response
    data = response.json()
    
    # If no more data, break loop
    if len(data['value']) == 0:
        break
    
    # Append data to list
    all_data.extend(data['value'])
    print(f"Fetched records: {skip} to {skip + len(data['value'])}")
    
    # Move to next batch
    skip += batch_size

# Convert to DataFrame
df_afstemning = pd.DataFrame(all_data)
print(f"Total records retrieved: {len(df_afstemning)}")
```

```python
df_afstemning.to_csv("./data/Denmark/dansk_data_afstemning.csv")
```

```python
import pandas as pd
from datetime import datetime

# Load your dataframe
# Assuming your data is stored in a file called 'political_data.csv'
# If it's stored elsewhere, modify the path accordingly
df = pd.read_csv("./data/Denmark/Merged.csv")

# Convert the dato column to datetime
# Based on your sample, the format appears to be 'YYYY-MM-DDThh:mm:ss'
df['dato'] = pd.to_datetime(df['dato'])

# Define date ranges
date_ranges = [
    ('1998-03-12', '2001-11-20', './data/Denmark/period_2001_2005.csv'),
    ('2001-11-21', '2005-02-08', './data/Denmark/period_2001_2005.csv'),
    ('2005-02-09', '2007-11-12', './data/Denmark/period_2005_2007.csv'),
    ('2007-11-14', '2011-09-14', './data/Denmark/period_2007_2011.csv'),
    ('2011-09-15', '2015-06-18', './data/Denmark/period_2011_2015.csv'),
    ('2015-06-19', '2019-06-05', './data/Denmark/period_2015_2019.csv'),
    ('2019-06-06', '2022-11-01', './data/Denmark/period_2019_2022.csv'),
    ('2022-11-02', pd.Timestamp.now().strftime('%Y-%m-%d'), './data/Denmark/period_2022_present.csv')
]

# Filter and save each period
for start_date, end_date, filename in date_ranges:
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)
    
    # Filter the dataframe
    period_df = df[(df['dato'] >= start) & (df['dato'] <= end)]
    
    # If any data exists for this period, save it
    if not period_df.empty:
        period_df.to_csv(filename, index=False)
        print(f"Saved {len(period_df)} rows to {filename}")
    else:
        print(f"No data found for period {start_date} to {end_date}")

print("Filtering and saving complete!")
```
